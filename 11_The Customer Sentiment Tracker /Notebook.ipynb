# The Customer Sentiment Tracker 

```python
# Reviews Analyzer using Transformer pipeline 
from transformers import pipeline
import pandas as pd

# Load pre-trained sentiment-analysis pipeline
sentiment = pipeline(
                    task="sentiment-analysis", 
                    model="distilbert-base-uncased-finetuned-sst-2-english", 
                    framework="pt"
)

input_reviews = [
    "Amazing quality, I am really happy with this purchase.",
    "Terrible experience, the product stopped working in two days.",
    "Not bad overall, but there is room for improvement.",
    "I am extremely satisfied, worth every penny.",
    "Very disappointing, I expected much better.",
    "Decent product, does the job but nothing special.",
    "Absolutely fantastic, exceeded my expectations.",
    "Worst product I have bought this year.",
    "Pretty good value for money, I like it.",
    "Completely useless, I regret buying this."
]

# Analyze sentiments
results = sentiment(input_reviews)
results_df = pd.DataFrame(results)
print(results_df)
```
